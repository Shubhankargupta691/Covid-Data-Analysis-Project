Now DataType of a  ' Death ' column  is string and replacing # with space

```python
covid_df.Death.str.replace('#','')
```

Therefore # is removed and now changing the Data type of " Death " str to int as we can't do calculation if it is in String/object

```python
covid_df['Death'] = covid_df.Death.str.replace('#','').astype(int)
```

```python
covid_df.dtypes
```

```python
C = (covid_df['Total_Confirmed_cases'] + covid_df['New_cases'])
L = (covid_df['Death']+covid_df['New_deaths']+covid_df['Cured']+covid_df['New_recovered'])
```

```python
covid_df['Active_cases'] = C - L
```

```python
State_wise = pd.pivot_table(covid_df, values = ["Total_Confirmed_cases", "Death", "Cured"],index="State/UT", aggfunc = max)
```

```python
State_wise["Recovery Rate"] = State_wise["Cured"]*100/State_wise["Total_Confirmed_cases"]
```

```python
State_wise["Mortality Rate"] = (State_wise["Death"])*100/State_wise["Total_Confirmed_cases"]
```

```python
State_wise = State_wise.sort_values(by = "Total_Confirmed_cases" , ascending= False)
```

```python
State_wise.style.background_gradient(cmap= "cubehelix")
```

Top 10 Active cases States

```python
Top_10_active_cases = covid_df.groupby(by = 'State/UT').max()[['Active_cases','Date']].sort_values(by = ['Active_cases'] , ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
ax = sns.barplot(data= Top_10_active_cases.iloc[:10], y = 'Active_cases', x = 'State/UT', linewidth = 2  ,edgecolor = 'black')
```

```python
Top_10_active_cases = covid_df.groupby(by = 'State/UT').max()[['Active_cases','Date']].sort_values(by = ['Active_cases'] , ascending=False).reset_index()

fig = plt.figure(figsize=(16,9))

ax = sns.barplot(data= Top_10_active_cases.iloc[:10], y = 'Active_cases', x = 'State/UT', linewidth = 2  ,edgecolor = 'black')

plt.title("Top 10 Active cases States in India" , size = 25)
plt.xlabel("Ststes")
plt.ylabel("Total Active cases")
plt.show()
```

Top States with highest number of death

```python
Top_10_deaths = covid_df.groupby(by='State/UT').max()[['Death','Date']].sort_values(by=['Death'],ascending=False).reset_index()

fig = plt.figure(figsize=(18,5))

plt.title("Top 10 States with most Deaths" , size = 25)

ax = sns.barplot(data= Top_10_deaths.iloc[:12], y = 'Death', x = 'State/UT', linewidth = 2  ,edgecolor = 'black')

plt.xlabel("States")
plt.ylabel("Total Death cases")
plt.show()

```
